```python
import pandas as pd
import requests

from openpyxl import load_workbook
from openpyxl import Workbook
```

```python
#url = 'https://www.worldfootball.net/schedule/eng-premier-league-2022-2023-spieltag/1/'
#url2 = 'https://www.worldfootball.net/schedule/eng-premier-league-2022-2023-spieltag/{x}/'

x = 1
y = str(x)

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

df_concat = []
df_concat = pd.DataFrame(df_concat)

while x <= 38:
  url = f'https://www.worldfootball.net/schedule/eng-premier-league-2022-2023-spieltag/{y}/'
  r = requests.get(url, headers=header)
  dfs = pd.read_html(r.text)
  df = dfs[1]
  df_url = df[[0, 1, 2, 4, 5]]
  df_url = pd.DataFrame(df_url)
  df_concat = pd.concat([df_concat, df_url])
  x = x + 1
  y = str(x)

tabla = df_concat

print(tabla)
```

```python
print(df_concat.shape)
print(df_concat.ndim)

x = 1
y = str(x)

tabla_sorted = []
tabla_sorted = pd.DataFrame(tabla_sorted)

while x <= 38:
  url = f'https://www.worldfootball.net/schedule/eng-premier-league-2022-2023-spieltag/{y}/'
  r = requests.get(url, headers=header)
  dfs = pd.read_html(r.text)
  df = dfs[1]
  sort_cols = df[[0,1,2,5,4]]
  sort_cols = pd.DataFrame(sort_cols)
  tabla_sorted = pd.concat([tabla_sorted, sort_cols])
  x = x + 1
  y = str(x)

tabla_sorted = tabla_sorted.fillna(axis=0, method='ffill').fillna(0)

print(tabla_sorted)

tabla_sorted.to_excel('fixturesTEST.xlsx', sheet_name='Fixtures')
```
